# Script for colocalization heatmap analysis of fluorescent intensity


```python
from PIL import Image, ImageOps, ImageEnhance, ImageFilter
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import seaborn as sns
```

```python
#Chose colour map in https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

#Functions
def Heatmap(image, merge, thresmin, thresmax):
    #Use image in RGB mode
    plt.figure(figsize = (8,8))
    plt.imshow(image)
    plt.show()

    if 'Green-Red' in merge:
        imageinput = cv2.bitwise_and(image[:,:,1], image[:,:,0])
        fig = plt.figure(figsize = (8,8))
        plot = plt.imshow(imageinput, cmap = 'hot', vmin = thresmin, vmax = thresmax)
        fig.colorbar(plot)

        colocalization = 0
        for row in range(imageinput.shape[0]):
          for col in range(imageinput.shape[1]):
            if thresmax > imageinput[row,col] > thresmin:
              colocalization = colocalization + imageinput[row,col]
        da = {'Colocalization signal per image': [colocalization]}
        dataframe = pd.DataFrame(data = da)

    if 'Green-Blue' in merge:
        imageinput = cv2.bitwise_and(image[:,:,1], image[:,:,2])
        fig = plt.figure(figsize = (8,8))
        plot = plt.imshow(imageinput, cmap = 'hot', vmin = thresmin, vmax = thresmax)
        fig.colorbar(plot)

        colocalization = 0
        for row in range(imageinput.shape[0]):
          for col in range(imageinput.shape[1]):
            if thresmax > imageinput[row,col] > thresmin:
              colocalization = colocalization + imageinput[row,col]
        da = {'Colocalization signal per image': [colocalization]}
        dataframe = pd.DataFrame(data = da)

    if 'Red-Blue' in merge:
        imageinput = cv2.bitwise_and(image[:,:,2], image[:,:,0])
        fig = plt.figure(figsize = (8,8))
        plot = plt.imshow(imageinput, cmap = 'hot', vmin = thresmin, vmax = thresmax)
        fig.colorbar(plot)

        colocalization = 0
        for row in range(imageinput.shape[0]):
          for col in range(imageinput.shape[1]):
            if thresmax > imageinput[row,col] > thresmin:
              colocalization = colocalization + imageinput[row,col]
        da = {'Colocalization signal per image': [colocalization]}
        dataframe = pd.DataFrame(data = da)

    return dataframe
```

```python
# OD TEMPORAL #

#Open and read image
path = r"/content/drive/MyDrive/Python protocols/Imagenes/Colocalization/Phall, Parkin y TOM20/SHAM T.tif"
image = cv2.imread(path)

#Show image in RGB
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (8,8))
plt.imshow(imageRGB)
plt.show()
imageRGB.shape

#HEATMAP ANALYSIS (RED GREEN CHANNELS)
minlvl = 10
maxlvl = 80

heatmap = Heatmap(imageRGB,'Green-Red', minlvl, maxlvl)
heatmap
```
